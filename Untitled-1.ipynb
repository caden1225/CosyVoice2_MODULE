```python
# 定义一个文本生成器（模拟实时文本到达）
def text_generator():
    """模拟实时文本流，例如从STT服务接收"""
    texts = ["我喜欢", "在", "周末", "听音乐"]
    for text in texts:
        print(f"[Text arrived] {text}")
        yield text
        time.sleep(0.5)  # 模拟网络延迟


# 文本流式输入 + 输出流式
for output in model.inference_sft(
    tts_text=text_generator(),  # ⭐ 传入Generator
    spk_id='caden_zh',
    stream=True,  # 同时启用输出流式
):
    tts_speech = output['tts_speech']
    audio_data = tts_speech.squeeze().cpu().numpy()
    display(Audio(audio_data, rate=model.sample_rate))
```
